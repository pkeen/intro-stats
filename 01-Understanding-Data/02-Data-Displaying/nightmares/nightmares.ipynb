```python
import pandas as pd

nightmares = pd.read_csv("../../data/nightmares.txt", sep="\t")

```

```python
nightmares_crosstab = pd.crosstab(nightmares["Side"], nightmares["Dream"])
print(nightmares_crosstab)
```

```python
nightmares['Dream'] = nightmares['Dream'].map({'N': 'Nightmare', 'n': 'No Nightmare'})
print(nightmares.head())
table = pd.crosstab(nightmares['Dream'], nightmares['Side'], margins=True)
print(table)
```

## Mosaic Plots
Mosaic plots are a type of data visualization that are used to display the relationship between two categorical variables. They are similar to contingency tables, but they are easier to interpret and can be used to display more complex relationships between variables.

```python
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt

# Mosaic plot
mosaic(nightmares, ['Side', 'Dream'])
plt.title('Mosaic Plot of Side vs Dream Outcome')
plt.show()
```
