```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/ceo-compensation-2018.txt", sep="\\t")
print(df.describe())
```

# Boxplot

```python
import seaborn as sns
from matplotlib.ticker import FuncFormatter

fig, ax = plt.subplots(figsize=(2, 6)) 

sns.boxplot(y="Compensation", data=df, ax=ax, width=0.5)
# Format y-axis ticks to show in millions (e.g. 120M)
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x*1e-6:.0f}M'))

ax.set_title("CEO Compensation")
ax.set_ylabel("Compensation ($M)")
plt.tight_layout()

```

# Histogram

```python
fig, ax = plt.subplots()


min_compensation = df["Compensation"].min()
max_compensation = df["Compensation"].max()

bins = range(0, max_compensation + 4_000_000, 2_000_000)

sns.histplot(data=df, x="Compensation", bins=bins, ax=ax)

ax.set_title("Distribution of CEO Compensation")
ax.set_xlabel("Compensation ($)")
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x*1e-6:.0f}M'))
ax.set_ylabel("# of CEOs")
plt.tight_layout()
```
