# Boxplots

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

cups = pd.read_csv('../data/cups.txt', sep='\t')

# 5 number summary
summary = cups.groupby('Container').describe()
print(summary)
```

```python
# Example using the built-in "tips" dataset
# tips = sns.load_dataset("tips")

# # Single variable
# sns.boxplot(x=tips["total_bill"])
# plt.title("Distribution of Total Bill")
# plt.show()

# Comparing a variable across groups
sns.boxplot(x="Container", y="Difference", data=cups)
plt.title("Temp lost by cup")
plt.xlabel("Cup type")
plt.ylabel("Temperature difference (F)")
plt.show()
```
