```python
# import the data 
import pandas as pd

df = pd.read_csv("../../data/framingham.txt", sep="\t")
df.head()
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
plt.xlabel("Diastolic BP (mm HG)")
plt.ylabel("Systolic BP (mm HG)")
plt.scatter(df["DBP"], df["SBP"])
plt.show()
```

```python
# Calculate correlation coefficient
import numpy as np

correlation = np.corrcoef(df["DBP"], df["SBP"])[0, 1]
print("Correlation coefficient:", correlation)
```
