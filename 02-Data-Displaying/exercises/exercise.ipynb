## Relative Frequency Chart

Create a relative frequency chart for the following data:

Live comfortably - 289
Meet their basic expenses with a little left over for extras - 2657
Just meet their basic expenses - 3297
Don’t even have enough to meet basic expenses - 594
Don’t know/No answer - 41



```python
import matplotlib.pyplot as plt

situation = [
    "Comfortable",
    "A little left over",
    "Basic expenses only",
    "Not enough for basic expenses",
    "Don't know/No answer"
]

frequency = [289, 2657, 3297, 594, 41]

# Increase figure size
plt.figure(figsize=(10, 6
))

# Create bar chart
plt.bar(situation, frequency)

# Rotate x-axis labels
plt.xticks(rotation=30, ha='right')

# Add labels for clarity (optional)
plt.ylabel('Number of People')
plt.title('Financial Situation Summary')

plt.tight_layout()  # Ensures everything fits
plt.show()
```

```python
# Change to relative frequency chart

age_categories = [
    "18-24",
    "25-25",
    "35-44",
    "45-54",
    "55-64",
    "65+"
]

frequency = [25049, 70764, 63397, 44698, 32516, 5041] 
total = sum(frequency)
relative_frequency = [f / total for f in frequency]  
# Optional: convert to percentages
percentages = [f * 100 for f in relative_frequency]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(age_categories, percentages)

# Add labels and rotate ticks
plt.ylabel('Percentage of Total (%)')
plt.xlabel('Age Category')
plt.title('Age Distribution as Percentage of Total')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
    
```
